```python
from tldraw import TldrawMatplotlib
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(6, 2))
ax.plot([0,10], [0,20], lw = 5);
plt.close()

widget = TldrawMatplotlib(my_figure=fig)
widget
```

```python
ax.plot([4,6], [20,0], lw = 7)

widget.update_plot(fig)
```
